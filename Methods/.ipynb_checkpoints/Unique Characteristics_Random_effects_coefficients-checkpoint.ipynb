```python
import os
os.chdir("..")

import time
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from jenkspy import jenks_breaks
from sklearn.cluster import KMeans
from kneed import KneeLocator
from sklearn.decomposition import PCA

from Thesis_utils.Thesis_models import *
from Thesis_utils import Thesis_viz as viz
from Thesis_utils import synthetic_data_generation as sd
```

```python
def run_model(model, all_args):
    arg_names = model.__code__.co_varnames[:model.__code__.co_argcount]
    model_args = {k: all_args[k] for k in arg_names if k in all_args}
    return model(**model_args)

def get_reduced_data(jenks_df, column_break):
    
    zone_indices = {zone : list(jenks_df[jenks_df[column_break] == zone].index) \
                                                for zone in np.unique(jenks_df[column_break])}
    mapping_dict = {value: key for key, values in zone_indices.items() for value in values}
    data = df.copy()
    data['gE_new'] = df['gV'].map(mapping_dict)
    
    return data
```

```python
seed = 50
mode = 'both'
df,_ = sd.create_data_linspace(n = 1000, 
                                n_effective_groups = 5, 
                                n_visible_groups = 200, 
                                n_contineous_features = 5, 
                                mode=mode, 
                                shuffle_groups=False,
                                re_error=(0, 0),
                                random_seed=seed,
                                show_random_effects=False)
np.random.seed(seed)
df.y = df.y + np.random.normal(0,3,len(df))
df
```

```python
print(MixedLM(df,to_drop='gV',mode='both')[0])
print(LMMNN(df,to_drop='gV')[0])
print(MERForest(df,to_drop='gV')[0])
print(ARMED(df,to_drop='gV')[0])
```

```python
# [MixedLM, LMMNN, LinearOHE, MERForest, ARMED]
Model = MERForest
```

```python
mse_with_Effective_groups = run_model(Model, {'df': df,'mode':mode,'to_drop':'gV'})[0]
mse_with_Visible_groups = run_model(Model, {'df': df,'mode':mode,'to_drop':'gE'})[0]
print("MSE_Effective_groups: ", mse_with_Effective_groups)
print("MSE_Visible_groups: ", mse_with_Visible_groups)
```

```python
start = time.time()
```

```python
md = sm.MixedLM.from_formula(f"y ~ {' + '.join([i for i in df.columns if i.startswith('f')])} + (1|gV)" \
                             ,df, groups = 'gV', re_formula='~'+'+'.join([i for i in df.columns if i.startswith('f')]))

mdf = md.fit()
mdf_re = pd.DataFrame([])
    
try:
    est_intercept = [((i+1)*mdf.fe_params['1 | gV']) + mdf.random_effects[i]['gV'] + mdf.fe_params[0] for i in mdf.random_effects.keys()]
    mdf_re['intercept'] = est_intercept

    for idx, f_i in enumerate([i for i in df.columns if i.startswith('f')]):
        
        est_slope = [mdf.fe_params[str(f_i)] + mdf.random_effects[i][str(f_i)] for i in range(len(mdf.random_effects))]
        mdf_re['slope_'+str(f_i)] = est_slope

except IndexError: pass
mdf_re
```

```python
X = np.array(mdf_re)

pca = PCA(n_components=1)
pca.fit(X)
X_transformed = pca.transform(X)
pca_df = pd.DataFrame(X_transformed, columns = [f'pca' for i in range(X_transformed.shape[1])])
```

```python
for i in range(2,11):
    
    array = np.array(X_transformed).reshape(len(X_transformed))
    bins = i
    try:
        pca_df['Jenks-'+str(bins)] = pd.cut(array,bins=jenks_breaks(array, bins), labels=range(bins) ,include_lowest=True, duplicates='drop')
    except ValueError: pass
pca_df
```

```python
res = pd.DataFrame([])

for itr, i in enumerate([col for col in pca_df.columns if col.startswith('J')]):
       
    reduced_df = get_reduced_data(pca_df, i)
    reduced_df.drop(['gE','gV'], axis=1, inplace=True)
    
    try:
    
#         train_mixedlm, test_mixedlm, _,_,_  = MixedLM_train_test(reduced_df, mode=mode)
#         train_lmmnn,   test_lmmnn,   _,_,_  = LMMNN_train_test(reduced_df)
        train_mixedlm, test_mixedlm,    _,_,_  = MERForest_train_test(reduced_df)
#         train_armed,   test_armed,   _,_,_  = ARMED_train_test(reduced_df)

        res.at[itr, 'name'] = i
        res.at[itr, 'train_mixedlm'] = train_mixedlm
        res.at[itr, 'test_mixedlm'] = test_mixedlm
#         res.at[itr, 'train_lmmnn'] = train_lmmnn
#         res.at[itr, 'test_lmmnn'] = test_lmmnn
#         res.at[itr, 'train_merf'] = train_merf
#         res.at[itr, 'test_merf'] = test_merf
#         res.at[itr, 'train_armed'] = train_armed
#         res.at[itr, 'test_armed'] = test_armed
        
    except np.linalg.LinAlgError: pass
```

```python
res
```

```python
fig, ax = plt.subplots(figsize = (7,5), dpi = 100)
results = res.copy()
results.reset_index(drop=True, inplace=True)

kn = KneeLocator(results.index, results['test_mixedlm'], curve='convex', direction='decreasing')
try:
    plt.axvline(x=results[results.test_mixedlm == kn.knee_y].index[0], color='k', linestyle='--', label='Knee Locator')
except ValueError: pass

# ax.plot(results.name, results.train_mixedlm, label='Train data')
ax.plot(results.name, results.test_mixedlm,  label='Test data')

plt.axline((0, mse_with_Visible_groups), (1, mse_with_Visible_groups), color = 'r', ls=':', linewidth = 2, label = 'MSE(before) - Visible Groups')

plt.xticks(rotation=60)
ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=12)
ax.legend()
ax.set_title("Jenks Natural Breaks", fontsize = 16, y = 1.02)
ax.set_xlabel('Jenks Break Number\n(N Clusters)', fontsize = 14)
ax.set_ylabel("Mean Squared Error (MSE)", fontsize = 14)
plt.tight_layout(pad = 3)
# plt.savefig(r"D:\Notes\Sem - 6\Thesis\Mixed Effect Model\Lit\Notebooks\Report\Methods\MixedLM_Jenks_merf.pdf",format = 'pdf',bbox_inches='tight')
```

```python
end = time.time()
print("Time Took:",np.round((end-start),2))
```

# KMeans


```python
X = np.array(mdf_re)
pca = PCA(n_components=2)
pca.fit(X)
X_transformed = pca.transform(X)
pca_df = pd.DataFrame(X_transformed, columns = [f'pca_'+str(i) for i in range(X_transformed.shape[1])])
X = np.array(pca_df)

K_range = range(2, 11)

for k in K_range:
    
    kmeans = KMeans(n_clusters= k, random_state=42)
    labels = kmeans.fit_predict(X)
    pca_df['K-'+str(k)] = labels
```

```python
res = pd.DataFrame([])

for itr, i in enumerate([col for col in pca_df.columns if col.startswith('K')]):
       
    reduced_df = get_reduced_data(pca_df, i)
    reduced_df.drop(['gE','gV'], axis=1, inplace=True)
    
    try:
    
#         train_mixedlm, test_mixedlm, _,_,_  = MixedLM_train_test(reduced_df, mode=mode)
#         train_lmmnn,   test_lmmnn,   _,_,_  = LMMNN_train_test(reduced_df)
        train_mixedlm, test_mixedlm,    _,_,_  = MERForest_train_test(reduced_df)
#         train_armed,   test_armed,   _,_,_  = ARMED_train_test(reduced_df)

        res.at[itr, 'name'] = i
        res.at[itr, 'train_mixedlm'] = train_mixedlm
        res.at[itr, 'test_mixedlm'] = test_mixedlm
#         res.at[itr, 'train_lmmnn'] = train_lmmnn
#         res.at[itr, 'test_lmmnn'] = test_lmmnn
#         res.at[itr, 'train_merf'] = train_merf
#         res.at[itr, 'test_merf'] = test_merf
#         res.at[itr, 'train_armed'] = train_armed
#         res.at[itr, 'test_armed'] = test_armed
        
    except np.linalg.LinAlgError: pass
```

```python
res
```

```python
fig, ax = plt.subplots(figsize = (7,5), dpi = 100)
results = res.copy()
results.reset_index(drop=True, inplace=True)

kn = KneeLocator(results.index, results['test_mixedlm'], curve='convex', direction='decreasing')
try:
    plt.axvline(x=results[results.test_mixedlm == kn.knee_y].index[0], color='k', linestyle='--', label='Knee Locator')
except ValueError: pass

# ax.plot(results.name, results.train_mixedlm, label='Train data')
ax.plot(results.name, results.test_mixedlm,  label='Test data')

plt.axline((0, mse_with_Visible_groups), (1, mse_with_Visible_groups), color = 'r', ls=':', linewidth = 2, label = 'MSE(before) - Visible Groups')

ax.tick_params(axis='x', labelsize=12)
ax.tick_params(axis='y', labelsize=12)
ax.legend()
ax.set_title("K-Means Clustering", fontsize = 16, y = 1.02)
ax.set_xlabel('\nN Clusters', fontsize = 14)
ax.set_ylabel("Mean Squared Error (MSE)", fontsize = 14)
plt.tight_layout(pad = 2)
# plt.savefig(r"D:\Notes\Sem - 6\Thesis\Mixed Effect Model\Lit\Notebooks\Report\Methods\MixedLM_KMeans_merf.pdf",format = 'pdf',bbox_inches='tight')
```
